# 1. 큐(Queue)

- 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 구조
- 줄을 서는 행위와 유사
- FIFO(First-in, First-Out)

### 1-1. 큐의 사용
- 멀티테스킹을 위한 프로세스 스케쥴링 방식을 구현(운영체제)
- 푸시메세지(문자)

### 1-2. 큐의 용어
- Enqueue : 큐에 데이터를 넣는 기능
- Dequeue : 큐에 데이터를 꺼내는 기능
- [VisualGo](https://visualgo.net/en/list)

### 1-3. queue 라이브러리를 활용
- Queue() : 가장 일반적인 큐 자료구조를 생성
- LifoQueue() : 나중에 입력된 데이터가 먼저 출력되는 구조
- PriorityQueue() : 데이터마다 우선순위를 넣어서 우선순위가 높은 순으로 출력


```python
# 일반적인 Queue 만들기
import queue

data_queue = queue.Queue()
data_queue.put('류호근') # Enqueue
print(data_queue)

print(data_queue.qsize())

data_queue.put('류포근') # Enqueue
print(data_queue.qsize())

print(data_queue.get()) # Dequeue
print(data_queue.qsize())
print(data_queue.get()) # Dequeue
print(data_queue.qsize())
```

```python
# 원시적인 큐는 같은 자료형만 넣을 수 있음
data_queue.put('류호근')
data_queue.put(22)
data_queue.put(True)
data_queue.put('내일 주말!')
print(data_queue.qsize())
```

```python
for i in range(data_queue.qsize()):
  item = data_queue.get()
  print(item)
print(data_queue.qsize())
```

```python
# LifoQueue 만들기
data_queue = queue.LifoQueue()
data_queue.put('류호근')
data_queue.put(22)
data_queue.put(True)
data_queue.put('내일 주말!')
print(data_queue.qsize())
print(data_queue.get())
```

```python
# PriorityQueue 만들기
# 값이 작을수록 우선순위가 높음
data_queue = queue.PriorityQueue()
# 튜플로 저장하여야 함
data_queue.put((10, '류호근'))
data_queue.put((14, '호근'))
data_queue.put((3, '류근'))
data_queue.put((16, '류포근'))
data_queue.put((9, '류호근헤헤'))
print(data_queue.qsize())
print(data_queue.get())
print(data_queue.get())
print(data_queue.get())
```

# 문제
리스트 변수로 Queue를 다루는 enqueue, dequeue 기능을 직접 구현해보자!

```python
queue_list = list() # []

def enqueue(data):
  queue_list.append(data)

def dequeue():
  data = queue_list[0]
  del queue_list[0]
  return data
```

```python
enqueue(10)
enqueue(5)
```

```python
print(dequeue())
```
